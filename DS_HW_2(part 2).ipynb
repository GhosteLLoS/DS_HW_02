```python
№9 Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
```

```python
№11 Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
```

```python
grouped_python_data = python_data.groupby("Посада")
```

```python
№12 Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.на.місяць"
```

```python
salary = grouped_python_data.agg({'Зарплата на місяць': ['min', 'max']})
salary
```

```python
№13 Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
```

```python
def fill_avg_salary(row):
    return row.mean()


salary[('Зарплата на місяць', 'avg')] = salary.apply(fill_avg_salary, axis=1)
salary
```

```python
№14 Створіть описову статистику за допомогою методу describe для нового стовпчика.
```

```python
stat = salary[('Зарплата на місяць', 'avg')].describe()
pd.DataFrame(stat.apply(lambda x: int(x)))
```
